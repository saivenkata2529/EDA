# working on outliers


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
dataframe = pd.read_csv('diabetes.csv')
dataframe.head(5)
```

```python
sns.histplot(data = dataframe['Age'], kde = True)
plt.axvline(x=dataframe.Age.mean(),color='red',alpha=0.5,label='Mean')
plt.axvline(x=dataframe.Age.median(),color='blue',ls='--',alpha=0.5,label='Median')
plt.legend()
```

```python
sns.boxplot(y=dataframe['Age'])
```

# Z-Score

```python
z = np.abs(stats.zscore(dataframe['Age']))
print(z)
```

```python
threshold = 3
print(np.where(z > 3)) # it the array it displays the 5 this are the outliers of the diabetes
```

```python
threshold = -3
print(np.where(z < -3))# their is no outliers
```

# IQR Score

```python
Q1 = np.percentile(dataframe['Age'], 25, interpolation = 'midpoint')
Q2 = np.percentile(dataframe['Age'], 50, interpolation = 'midpoint')
Q3 = np.percentile(dataframe['Age'], 75, interpolation = 'midpoint')
```

# Dealing with Skewed Data

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head(5)
```

```python
sns.histplot(data = df['price'], kde = True)
plt.axvline(x=df.price.mean(),color='red',alpha=0.5,label='Mean')
plt.axvline(x=df.price.median(),color='blue',ls='--',alpha=0.5,label='Median')
plt.legend()
```

# In the price data it shows the mean and median range of the data to the histogram plot it will describe the matter very clear to understand

# Data is transformed is the process of math function into the mean and medain the data goes is right skew it is right skewed data

```python
# create the log transformaded data
df['price_log'] = np.log(df['price'])
```

```python
sns.displot(df['price_log'])
```

```python
#Square root transformation
df['price_sqrt'] = np.sqrt(df['price'])
```

```python
sns.displot(df['price_sqrt'])
```

```python
#Power transformation
```

```python
from scipy.stats import boxcox

#Boxcox transformation in python
df.insert(len(df.columns), 'A_Boxcox',
          boxcox(df.iloc[:, 2])[0])
          
```

```python
sns.displot(df['A_Boxcox'])
```
